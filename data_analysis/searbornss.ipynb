```python
import seaborn as sns
```

```python
# basic plotting with seaborn 
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.scatterplot(tips, color= "red")
# import matplotlib as plt
# sns.scatterplot(x= 'total bill', y='tip', data =tips)
```

```python
# line plot  
sns.lineplot(tips)
```

```python
sns.barplot(tips)
```

```python
sns.pairplot(tips)
```

```python
# heatmap 
sns.heatmap(tips['total_bill','tip','size'].corr(), annot=True)
```
