```python
import matplotlib.pyplot as plt


```

```python
x = [1,2,3,4,5]
y = [2,4,6,18,10]
# create a line plot 
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,18,10]
# customzize line plot 
plt.plot(x, y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.grid(True)
plt.show()

# plt.show()
```

```python
# multiple plots  
# sample data 
x1 = [1,2,3,4,5]
y1 = [2,4,6,18,10]
x2 = [1,2,3,4,5]
y2 = [4,8,12,24,20]

# create two line plots
plt.plot(x1, y1, label='first line')
plt.plot(x2, y2, label='second line')
```

```python
x1 = [1,2,3,4,5]
y1 = [2,4,6,18,10]
x2 = [1,2,3,4,5]
y2 = [4,8,12,24,20]

plt.subplot(1,2,1)
plt.plot(x1, y1, label='first line')
plt.title('first line')

plt.subplot(1,2,2)
plt.plot(y1, x1, label='first line')
plt.title('first line')
plt.subplot(1,2,1)
plt.plot(x2, y2, label='second line')
plt.plot(y1, x1, label='first line')
plt.title('first line')
plt.subplot(1,2,2)
plt.plot(y1, x1, label='first line')
plt.title('first line')
```

```python
# bar plot  
categories = ["A", "B", "C", "D"]
values = [10, 20, 10, 40]
plt.bar(categories, values, color='black')
```

```python
# histogram  
data = [1,2,3,4,5,6,7,8,9]
plt.hist(data, bins=5, color='blue')
```

```python
# scatter plot  
x = [1,2,3,4,5]
y = [2,4,6,18,10]
plt.scatter(x, y, marker='x', color='red')
plt.show()

# pie chart
```

```python
# pie chart  
labels = 'Frogs', 'Cats', 'Dogs', 'Fish'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e., 'Cats')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
```
